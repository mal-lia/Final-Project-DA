# **Visualisasi Tabel**

Secara keseluruhan, grafik ini menunjukkan **konsentrasi peserta yang sangat tinggi di wilayah Jawa, mencerminkan distribusi geografis yang timpang dalam partisipasi kegiatan.**

```python
import pandas as pd
import matplotlib.pyplot as plt

gender_ratio_per_department = pd.crosstab(df['Departemen apa yang kamu inginkan?'], df['Jenis Kelamin'], normalize='index') * 100

# Scatter Plot Pembanding
plt.figure(figsize=(10, 6))
for department in gender_ratio_per_department.index:
  male_ratio = gender_ratio_per_department.loc[department, 'Laki-laki']
  female_ratio = gender_ratio_per_department.loc[department, 'Perempuan']
  plt.scatter(male_ratio, female_ratio, label=department)

plt.xlabel("Rasio Pria (%)")
plt.ylabel("Rasio Wanita (%)")
plt.title("Perbandingan Pria vs Wanita per Departemen")
plt.legend()
plt.grid(True)
plt.show()
```

> **Analisis Scatterplot Perbandingan Gender per Departemen**

* Visualisasi ini menunjukkan perbandingan rasio pria dan wanita dalam setiap departemen dalam bentuk scatter plot. Setiap titik merepresentasikan satu departemen, dengan sumbu X menunjukkan rasio pria (%) dan sumbu Y menunjukkan rasio wanita (%). Terlihat bahwa **sebagian besar departemen didominasi oleh peserta wanita, seperti Video Editor (100% wanita), Graphic Designer, Administration and Finance, dan Social Media Officer, yang semuanya memiliki rasio wanita di atas 80%.**

* Beberapa departemen memiliki **distribusi yang lebih seimbang, seperti Project Officer, Public Relation, dan Human Resource and Development (HRD) dengan rasio pria sekitar 29–31%.** Sementara itu, Marketing dan terutama WordPress Developer cenderung lebih banyak diisi oleh peserta pria, dengan **WordPress Developer menjadi satu-satunya departemen yang didominasi oleh pria (hampir 60%).** Grafik ini mengindikasikan adanya kecenderungan gender tertentu dalam pemilihan departemen, di mana **peran administratif dan kreatif cenderung menarik lebih banyak wanita, sedangkan bidang teknis lebih banyak diminati pria.**

```python
import matplotlib.pyplot as plt
provinsi_labels = provinsi.index
provinsi_counts = provinsi.values

plt.figure(figsize=(12, 6))
plt.scatter(provinsi_labels, provinsi_counts, color = 'skyblue')
plt.xlabel("Provinsi")
plt.ylabel("Jumlah Peserta")
plt.title("Distribusi Peserta per Provinsi")
plt.xticks(rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

> **Analisis ScatterPlot Distribusi Peserta per Provinsi**

* Visualisasi ini menunjukkan distribusi jumlah peserta berdasarkan provinsi dalam bentuk scatter plot. **Setiap titik merepresentasikan satu provinsi, dengan posisi vertikal menunjukkan jumlah peserta dari provinsi tersebut.** Terlihat bahwa dominasi peserta berasal dari Jawa Barat, Jawa Timur, dan Jawa Tengah, yang jumlah pesertanya jauh lebih tinggi dibandingkan provinsi lain.

* Setelah **lima provinsi teratas**, seperti DKI Jakarta dan Banten, jumlah peserta menurun drastis. Sebagian besar provinsi lainnya, khususnya dari luar Pulau Jawa seperti di Kalimantan, Sulawesi, dan Nusa Tenggara, hanya menyumbang beberapa peserta saja. **Pola ini menunjukkan konsentrasi partisipasi yang sangat tinggi di wilayah Jawa, sedangkan keterlibatan dari daerah-daerah lain masih sangat rendah.**
